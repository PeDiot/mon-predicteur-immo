```python
def load_zip_dvf(zip_dir: str, year: int) -> DataFrame: 
    zf = zipfile.ZipFile(zip_dir) 
    file_name = f"dvf/{year}.csv"

    df = pd.read_csv(zf.open(file_name))
    return df 

df = load_zip_dvf(ZIP_DIR, year=2021)
```

### Cleaning de la base dvf21

```python
def select_sales(df: DataFrame) -> DataFrame: 
    if "nature_mutation" not in df.columns: 
        raise ValueError("nature_mutation not in columns.")
    
    return df.loc[df.nature_mutation == "Vente", :] 
df_sales = select_sales(df)
df_sales.head()
```

```python
print(f"{df_sales.shape[0]} transactions associées à des ventes \n{df_sales.id_mutation.unique().shape[0]} identifiants uniques de mutation")
```

```python
def remove_industrial_facilities(df: DataFrame) -> DataFrame: 
    if "code_type_local" not in df.columns: 
        raise ValueError("code_type_local not in columns.")
    
    return df.loc[
        (df.code_type_local >= 1) & 
        (df.code_type_local < 4), 
        :
    ] 
df_sales = remove_industrial_facilities(df_sales)
print(f"{df_sales.shape[0]} ventes d'appartements/maisons/dépendances")
```

```python
pd.crosstab(df_sales["type_local"], df_sales["code_type_local"], margins=True)
```

### Cleaning de la base dvf21

```python
# traitement des actes simples 
temp = df_sales.groupby("id_mutation").size()
not_duplicated = temp[temp==1].index.tolist()
not_duplicated
```

```python
mutations_simples = df_sales.loc[df_sales.id_mutation.isin(not_duplicated), :]
mutations_simples.head()
```

```python
temp = df_sales.groupby("id_mutation").size()
duplica = temp[temp>1].index.tolist()
duplica
```

```python
mutations_complexes = df_sales.loc[df_sales.id_mutation.isin(duplica), :]
mutations_complexes.head()
```

```python
# Application de la solution 4 sur les mutations dites complexes (cf. Coulondre, 2019)

var_quali = ['type_local', 'nature_mutation'] # add more here
for var in var_quali: 
  clean_quali = mutations_complexes.groupby(['id_mutation', 'valeur_fonciere'])[f'{var}'].apply(lambda x: ','.join(x)).reset_index()

var_quanti = ['nombre_lots','surface_reelle_bati'] # add more here
for var in var_quanti:
  clean_quanti = mutations_complexes.groupby(['id_mutation', 'valeur_fonciere'])[f'{var}'].sum()
```

```python
# Select columns and merge in a df 
mutations_complexes_clean = pd.merge(clean_quanti, clean_quali, on=['id_mutation'])
mutations_complexes_clean.head(10)
```

```python
# Merge/append again with no_duplica since duplica has been treated
df_mutation_clean = mutations_simples.append(mutations_complexes_clean)
df_mutation_clean.head(10)
```
